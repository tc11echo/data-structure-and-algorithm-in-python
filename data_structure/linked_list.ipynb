# Single Linked List

linked-list-concept.webp

```python
class Node:
  def __init__(self, data, next=None):
    self.data=data
    self.next=next

  def __repr__(self):
    if self:
      return f"{self.data} -> {self.next}"

class SingleLinkedList:
  def __init__(self):
    self.head=None

  def __repr__(self):
    return f'{self.head}'

  def empty(self):
    return self.head is None

  def get_size(self):
    size=0
    cur=self.head
    while cur: # while cur is not None
      # when cur is not None, then bool(cur)=true
      size+=1
      cur=cur.next
    return size 

  def get_data(self, index):
    # 0<=i<=size-1
    count=0 # show the index of the node I am now in
    cur=self.head
    while cur and count!=index:
      cur=cur.next
      count+=1
    if cur is None: # the i is out of bound
      return None
    else: # the i is in-bound
      return cur.data

  def index_of_by_node(self, node):
    # node is a Node, not a data
    count=0
    cur=self.head
    while cur and cur is not node:
      cur=cur.next
      count+=1
    # when we exit while loop, there are two possibilities:
    # 1. cur == None => node is not in the linked list
    # 2. cur != None => node points to node
    if cur is None:
      return "no such node"
    else:
      return count

  def index_of_by_data(self, data):
    count=0
    cur=self.head
    while cur and cur.data!=data:
      cur=cur.next
      count+=1
    # when we exit while loop, there are two possibilities:
    # 1. cur==None => el is not in the linked list
    # 2. cur!=None => el points to el
    if cur is None:
      return "no such data"
    else:
      return count

  def insert_head(self, data):
    # data is a value, not a Node
    # we need to create a Node
    new=Node(data, next=self.head)
    self.head=new

  def insert_tail(self, data):
    new=Node(data)
    cur=self.head
    if self.empty(): #if cur is None: (the list is empty)
      self.head=new
      return
    while cur.next:
      cur=cur.next
    cur.next=new

  def insert_by_index(self, index, data):
    if index==0:
      self.insert_head(data)
    else:
      count=0 # want count to become index-1
      cur=self.head
      while cur and count!=index-1:
        cur=cur.next
        count+=1 
      # when we exit while loop, two possibilities
      # 1. cur is None
      # 2. count == index-1
      if cur is None:
        print("index out of bound...")
      else:
        cur.next=Node(data, next=cur.next)
     
  def remove_by_index(self, index):
    if index==0:
      self.head=self.head.next
    else:
      cur=self.head
      count=0 # want count to become index-1
      while cur and count!=index-1:
        cur=cur.next
        count+=1 
      # when we exit while loop, two possibilities
      # 1. cur is None
      # 2. count == index-1
      # print(cur.data)
      if cur is None or cur.next is None:
        print("index out of bound...")
      else:
        cur.next=cur.next.next

  def remove_by_data(self, data):
    cur=self.head
    if cur is not None:
      #remove the first node
      if cur.data==data:
        self.head=cur.next
        return
    while cur is not None and cur.data!=data:
      prev=cur
      cur=cur.next
    #no match node is found
    if cur is None:
      return
    else:
      prev.next=cur.next #remove the node

  def display_all(self):
    if self.empty():
      print("stack is empty")
      return
      
    temp=""
    printval=self.head
    while printval:
      temp=temp+str(printval.data)+" -> "
      printval=printval.next
    return temp

    '''printval=self.head
    while printval:
      print(printval.data, end=" -> ")
      printval=printval.next'''

t=SingleLinkedList()
t.insert_tail(0)
t.insert_tail(1)
t.insert_tail(2)
t.insert_tail(3)
t.insert_tail(4)
t.insert_tail(5)
print(t)
print(t.index_of_by_data(4))
t.remove_by_index(1)
print(t)
t.insert_by_index(1,1)
print(t)
t.remove_by_data(5)
t.remove_by_data(3)
print(t)
t.insert_head(-1)
print(t)
print(t.get_data(3), t.index_of_by_data(2), t.empty(), t.get_size())
print(t.display_all())
```

# Doubly Linked List

doubly-linked-list-created.webp

```python
class Node:
  def __init__(self, data, prev=None, next=None):
    self.data=data
    self.prev=prev
    self.next=next

  def __repr__(self):
    if self:
      return f"{self.data} -> {self.next}"

class DoublyLinkedList:
  def __init__(self):
    self.head=None

  def __repr__(self):
    return f'{self.head}'

  def empty(self):
    return self.head is None
 
  def insert_head(self, data):
    new=Node(data, None, self.head)
    # change prev of head node to new
    if self.head is not None:
      self.head.prev=new
    # move the head to point to the new node
    self.head=new

  def insert_tail(self, data):
    new=Node(data)
    if self.head is None:
      self.head=new
      return
    cur=self.head
    while cur.next is not None:
      cur=cur.next
    cur.next=new
    new.prev=cur
    return

  def insert_after(self, cur_prev, data):
    # check if the given prev is None
    if cur_prev is None:
      print("the given previous node cannot be NULL")
      return
    new=Node(data, cur_prev, cur_prev.next)
    cur_prev.next=new
    if new.next is not None:
      new.next.prev=new

  def reverse(self):
    temp=None
    cur=self.head
    # Swap next and prev for all nodes of doubly linked list
    while cur is not None:
      temp=cur.prev
      cur.prev=cur.next
      cur.next=temp
      cur=cur.prev
    # Before changing head, check for the cases like empty list and list with only one node
    if temp is not None:
      self.head=temp.prev

  def display_all(self):
    printval=self.head
    print("forward direction: ", end="")
    while printval:
      print(printval.data, end=" -> ")
      temp=printval
      printval=printval.next
    print()
    print("reverse direction: ", end="")
    while temp:
      print(temp.data, end=" -> ")
      temp=temp.prev
    print()

d=DoublyLinkedList()
d.insert_tail(6)
d.insert_head(7)
d.insert_head(1)
d.insert_tail(4)
d.insert_after(d.head.next,8)
print(d)
d.display_all()
d.reverse()
#d.display_all()
print(d)
```
