# Hash
* a technique that determines the index of a key
* `hash()` in python

Hash_table.png

### Hash Function
* take a search key and produce the integer index of an element in the hash table
  * When applied to an object, return a number (index)
  * When applied to equal object, return the same number
  * When applied to unequal object, return different number (but it can still be likely to return the same number, this determent by the hash function)
* Preliminary examples of hash functions:
  * hash(x) = x/n where n is 11
  * hash(x) = x % no of buckets
  * prime number division
  * mid square
  * move or folding

### Hash Table
* array storing values indexed by hash function

### How Hashing Function and Hash Table Work Together?
* use an array of size $b$
  * Each position of this array is a bucket
* use a hash function $f$ that convert each key $k$ into an index in the range $[0,b-1]$
* every dictionary pair (key, value) is stored in its home bucket table[$f(key)$]

### Efficiency

* Hash tables are surprisingly efficient
* Until the table is about 70% full, the number of
probes (places looked at in the table) is typically
only 2 or 3
* Sophisticated mathematical analysis is required to
prove that the expected cost of inserting into a hash table, or looking something up in the hash table, is constant time
* Even if the table is nearly full (leading to occasional long searches), efficiency is usually still quite high

* Prefect hashing maps each search key into a different integer that is suitable as an index to hash table
* Search time can be only O(1)

# Collision
* When two values hash to the same array
location
* Collision are normally treated as "first come,
first served", the first value that hashes to the location gets it


### Handling collision
* Open Addressing
  * Linear Probing
    * Search from there for an empty location
  * Quadratic Probing
  * Double Hashing
* Closed Addressing
  * Separate Chaining
    * Use the array location as the header of a linked list of values that hash to this location


```python
hash_table_size=5
hash_table=[0]*hash_table_size

# use % mod as hash function
hash_table[(hash("apple")%hash_table_size)]="apple"
print(hash_table)
hash_table[(hash("graps")%hash_table_size)]="graps"
print(hash_table)
hash_table[(hash("banana")%hash_table_size)]="banana"
print(hash_table)
# still now it is stilll perfect hashing

# collision occur
hash_table[(hash("strawberry")%hash_table_size)]="strawberry"
print(hash_table)
```

* When a bucket can hold only one pair, collisions and overflows occur together
