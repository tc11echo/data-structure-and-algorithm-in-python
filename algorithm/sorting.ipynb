# Selection Sort

### Description
* Work by finding the smallest (or largest, depending on sorting order) element in the unsorted sublist, swapping it with the leftmost unsorted element (putting it in sorted order), and moving the sublist boundaries one element to the right.

### Type
* Algorithm Strategy : Brute Force
* Algorithm Category : Selection
* Comparative Sort
* In-place
* Not-stable
* Simple

#### Early-Terminating Selection Sort
* In Early-Terminating Selection Sort, during the scan for the largest element, it checks to see whether the array is already sorted 

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n^2)|O(n^2)|O(n^2)|O(1)|

<br/>

````python
# Selection Sort(find min ascending)
def selection_sort_min_ascending(arr):
  for i in range(len(arr)-1):
    min_index=i
    for j in range(i+1,len(arr)):
      if arr[j]<arr[min_index]:
        min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
selection_sort_min_ascending(arr)
print(arr)
print()


# Selection Sort(find min descending)
def selection_sort_min_descending(arr):
  for i in range(len(arr)-1):
    min_index=i
    for j in range(i+1,len(arr)):
      if arr[j]>arr[min_index]:
        min_index=j
    arr[i],arr[min_index]=arr[min_index],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
selection_sort_min_descending(arr)
print(arr)
print()


# Selection Sort(find max ascending)
def selection_sort_max_ascending(arr):
  for i in range(len(arr)-1,-1,-1):
    max_index=i
    for j in range(i-1,-1,-1):
      if arr[j]>arr[max_index]:
        max_index=j
    arr[i],arr[max_index]=arr[max_index],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
selection_sort_max_ascending(arr)
print(arr)
print()


# Selection Sort(find max descending)
def selection_sort_max_ascending(arr):
  for i in range(len(arr)-1,-1,-1):
    max_index=i
    for j in range(i-1,-1,-1):
      if arr[j]<arr[max_index]:
        max_index=j
    arr[i],arr[max_index]=arr[max_index],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
selection_sort_max_ascending(arr)
print(arr)
print()
````

# Bubble Sort

### Description
* Work by comparing the two adjacent elements. If the left element is smaller than the right element, then remains unchanged. If the left element is bigger than the right element, then swap them and compare the right element to the next adjacent element and repeat the above step.

### Type
* Algorithm Strategy : Brute Force
* Algorithm Category : Exchanging
* Comparative Sort
* In-place
* Stable
* Simple

#### Early-Terminating Bubble Sort
* If a bubbling pass results in no swaps, then the array is in sorted order and no further bubbling passes are necessary.
* Note: It needs a pass for the checking of order

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n)|O(n^2)|O(n^2)|O(1)|

<br/>

````python
# Bubble Sort(ascending)
def bubble_sort_ascending(arr):
  for j in range(len(arr)-1):
    for i in range(len(arr)-1-j):
      if arr[i]>arr[i+1]:
        arr[i],arr[i+1]=arr[i+1],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
bubble_sort_ascending(arr)
print(arr)
print()


# Bubble Sort(descending)
def bubble_sort_descending(arr):
  for j in range(len(arr)-1):
    for i in range(len(arr)-1,j,-1):
      if arr[i]>arr[i-1]:
        arr[i],arr[i-1]=arr[i-1],arr[i]

arr=[3,1,41,59,26,53,47]
print(arr)
bubble_sort_descending(arr)
print(arr)
print()


# Bubble Sort(for-loop early-terminating)
def bubble_sort_for_earlyterm(arr):
  for j in range(len(arr)-1):
    sorted=True
    for i in range(len(arr)-1-j):
      if arr[i]>arr[i+1]:
        sorted=False
        arr[i],arr[i+1]=arr[i+1],arr[i]
    if sorted: return
    
arr=[3,1,41,59,26,53,47]
print(arr)
bubble_sort_for_earlyterm(arr)
print(arr)
print()


# Bubble Sort(while-loop early-terminating)
def bubble_sort_while_earlyterm(arr):
  swapped=True
  j=0
  while(swapped):
    swapped=False
    for i in range(len(arr)-j-1):
      if arr[i]>arr[i+1]:
        swapped=True
        arr[i],arr[i+1]=arr[i+1],arr[i]
    j+=1
    
arr=[3,1,41,59,26,53,47]
print(arr)
bubble_sort_while_earlyterm(arr)
print(arr)
print()
````

# Cocktail Sort / Bidirectional Bubble Sort / Shaker Sort / Ripple Sort / Shuffle Sort / Shuttle Sort

### Description
* Work by extending bubble sort by operating in two directions
* Variance of bubble sort

### Type
* Algorithm Strategy : Brute Force
* Algorithm Category : Exchanging
* Comparative Sort
* Stable
* In-place

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n)|O(n^2)|O(n^2)|O(1)|

<br/>

````python
# Cocktail Sort
def cocktail_sort(arr):
  length=len(arr)
  for i in range(length, 0, -1):
    swapped=False
    cut_off=abs(i-length)
    #print(cut_off)
    obverse_count=len(arr[cut_off:i-1])
    reverse_count=len(arr[cut_off+1:i-1])
    
    for j in range(obverse_count):
      if arr[j]>arr[j+1]:
        swapped=True
        arr[j],arr[j+1]=arr[j+1],arr[j]
    #print(arr)
    for j in range(reverse_count, 0, -1):
      if arr[j]<arr[j-1]:
        swapped=True
        arr[j],arr[j-1]=arr[j-1],arr[j]
    #print(arr)
    if not swapped: return

arr=[6,5,4,3,2,100,9]
cocktail_sort(arr)
print(arr)
````

# Comb Sort
### Description
* Work by improving on bubble sort in the same way that shell sort improves on insertion sort
* Variance of bubble sort

### Type
* Algorithm Strategy : Brute Force
* Algorithm Category : Exchanging
* Comparative Sort
* Not-stable
* In-place

Shrink Factor=1.3

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nlogn)|O(n^2/2^p)|O(n^2)|O(1)|
where $p$ is the number of increment

<br/>

````python
"""# shrink gap by shrink factor
def get_next_gap(gap):
  gap=(gap*10)//13
  if gap<1: return 1
  return gap"""

from math import floor
# shrink gap by shrink factor
def get_next_gap(gap):
  SHRINK_FACTOR=1.3
  gap=floor(gap/SHRINK_FACTOR)
  if gap<1: return 1
  return gap

# Comb Sort
def comb_sort(arr):
  swapped=True
  length=len(arr)
  gap=length

  while gap!=1 or swapped is True:
    swapped=False
    gap=get_next_gap(gap)

    for i in range(0, length-gap):
      if arr[i]>arr[i+gap]:
        swapped=True
        arr[i],arr[i+gap]=arr[i+gap],arr[i]

arr=[6,5,4,3,2,100,9]
comb_sort(arr)
print(arr)
````

# Insertion Sort

### Description
* Work by consuming one input element each repetition, and grows a sorted output list. At each iteration, insertion sort removes one element from the input data, finds the location it belongs within the sorted list, and inserts it there. It repeats until no input elements remain.


### Type
* Algorithm Strategy : Brute Force
* Algorithm Category : Insertion
* Comparative Sort
* In-place
* Stable
* Simple

#### Early-Terminating Insertion Sort

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n)|O(n^2)|O(n^2)|O(1)|

<br/>

````python
# Insertion Sort (iterative)
def insertion_sort_iterative(arr):
  for i in range(1,len(arr)):
    key=arr[i]
    j=i-1
    while j>=0 and key<arr[j]:
      arr[j+1]=arr[j]
      j-=1
      # print("   ",end="")
      # print(arr)
    arr[j+1]=key
    # print(arr)

#arr=[13,24,14,16,60,39,8,17,50,10]
arr=[11,12,13,5,6,59,26,53,47]
print(arr)
insertion_sort_iterative(arr)
print(arr)
print()


# Insertion sort (recursive)
def insertion_sort_recursive(arr):
  if len(arr)<=1: # only one element
    return arr
  arr[:-1]=insertion_sort_recursive(arr[:-1])
  last_element=arr[len(arr)-1]
  i=len(arr)-2 # start from the last -1 element
  while i>=0 and arr[i]>last_element: # loop until the smallest, within arr[0, i] is in order
    arr[i+1]=arr[i]
    i-=1
  arr[i+1]=last_element
  return arr
    
arr=[19,13,6,2,18,8]
print(arr)
insertion_sort_recursive(arr)
print(arr)
print()
````

# Shell Sort

### Description
* Work by optimizating the insertion sort that allows the exchange of items that are far apart
* Variance of insertion sort

### Type
* Algorithm Strategy : 
* Algorithm Category : Insertion
* Comparative Sort
* In-place
* Not-stable
* Efficient

#### Early-Terminating Insertion Sort

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nlogn) (most gap sequences)|depends on gap sequence|O(n^2) (worst known worst case gap sequence)|(n) total, O(1) auxiliary|
|O(nloglogn) (best known worst-case gap sequence)||O(nloglogn) (best known worst case gap sequence)||

<br/>

````python
# Shell Sort
def shell_sort(arr):
  n=len(arr)
  gap=n//2
  while gap>0:
    for i in range(gap,n):
      temp=arr[i]
      j=i
      while j>=gap and arr[j-gap]>temp:
        arr[j]=arr[j-gap]
        j=j-gap
      arr[j]=temp
    gap=gap//2

arr=[89,34,23,78,67,100,66,29,79,55,78,88,92,96,96,23]
print(arr)
shell_sort(arr)
print(arr)
````

# Merge Sort

### Description
* Work by dividing the unsorted list into n sublists, each containing one element (a list of one element is considered sorted). Repeatedly merge sublists to produce new sorted sublists until there is only one sublist remaining. This will be the sorted list.
* middle=$\lfloor \frac{left+right}{2} \rfloor$=`(left+right)//2` in python


### Type
* Algorithm Strategy : Divide and Conquer
* Algorithm Category : Merging
* Comparative Sort
* Not-in-place
* Stable
* Efficient

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nlogn)|O(nlogn)|O(nlogn)|O(n)|

````python
# Merge Sort(one parameter)
def merge_sort_one_parameter(arr):
  if len(arr)>1:
    mid=len(arr)//2
    L=arr[:mid]
    R=arr[mid:]
    merge_sort_one_parameter(L)
    merge_sort_one_parameter(R)
    merge_one_parameter(arr,L,R)

def merge_one_parameter(arr,L,R):
  i,j,k=0,0,0
  # Copy data to temp arrays L[] and R[]
  while i<len(L) and j<len(R):
    # descending if L[i]>R[j]:
    if L[i]<R[j]:
      arr[k]=L[i]
      i+=1
    else:
      arr[k]=R[j]
      j+=1
    k+=1
  # Checking if any element was left
  while i<len(L):
    arr[k]=L[i]
    i+=1
    k+=1
  while j<len(R):
    arr[k]=R[j]
    j+=1
    k+=1

arr=[3,12,1,41,59,26,7,53,4]
print(arr)
merge_sort_one_parameter(arr)
print(arr)
print()


# Merge Sort(three parameter no len())
def merge_sort_three_parameter(arr, first, last):
  if first>=last:
    return arr
  if first<last:
    mid=first+(last-first)//2
    merge_sort_three_parameter(arr, first, mid)
    merge_sort_three_parameter(arr, mid+1, last)
    merge_three_parameter(arr, first, mid, last)

def merge_three_parameter(arr, first, mid, last):
  L=[]
  R=[]
  # remark
  n1=mid-first+1
  n2=last-mid

  # Copy data to temp arrays L[] and R[]
  for i in range(0,n1):
    L.append(arr[first+i])
  for i in range(0,n2):
    R.append(arr[mid+1+i])

  i,j,k=0,0,first
  # Copy data from temp arrays L[] and R[]
  while i<n1 and j<n2:
    # descending if L[i]>R[j]:
    if L[i]<R[j]:
      arr[k]=L[i]
      i+=1
    else:
      arr[k]=R[j]
      j+=1
    k+=1
  # Checking if any element was left
  while i<n1:
    arr[k]=L[i]
    i+=1
    k+=1
  while j<n2:
    arr[k]=R[j]
    j+=1
    k+=1

arr=[3,12,1,41,59,26,7,53,4]
print(arr)
merge_sort_three_parameter(arr,0,len(arr)-1)
print(arr)
print()
````

# Quick Sort

### Description
* Work by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot.


### Type
* Algorithm Strategy : Divide and Conquer
* Algorithm Category : Partitioning
* Comparative Sort
* In-place (Loose) /  Not-in-place (Strict)
* Not-stable
* Efficient

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nlogn)|O(nlogn)|O(n^2)|O(logn)|

<br/>

---

```
# Pseudocode:

partition(arr, low, high)
	pivot  arr[low]
  i  low
  for j  low+1 to high
    if arr[j]<pivot
      i  i+1
      swap(arr[i], arr[j])
    end if
  end for j
 	swap(arr[i], arr[low])
	return i

quick_sort(arr, low, high)
  if left >= right return
  else
    pivot  partition (arr, low, high)
    quick_sort (arr, low, pivot-1)
    quick_sort (arr, pivot+1, high)
  end if
```



````python
# Quick Sort(low)
def partition_low(arr, low, high):
  pivot=arr[low]
  i=low
  for j in range(low+1,high+1):
    if arr[j]<pivot:
      i+=1
      arr[i],arr[j]=arr[j],arr[i]
  arr[i],arr[low]=arr[low],arr[i]
  return i

def quick_sort_low(arr, low, high):
  if len(arr)==1:
    return arr
  # can use  if low>=high: return arr
  if low<high:
    pivot=partition_low(arr, low, high)
    quick_sort_low(arr, low, pivot-1)
    quick_sort_low(arr, pivot+1, high)

arr=[3,12,1,4,59,26,7,53,4]
print(arr)
quick_sort_low(arr,0,len(arr)-1)
print(arr)
print()


# Quick Sort(high)
def partition_high(arr, low, high):
  pivot=arr[high]
  i=low-1
  for j in range(low,high):
    if arr[j]<pivot:
      i+=1
      arr[i],arr[j]=arr[j],arr[i]
  arr[i+1],arr[high]=arr[high],arr[i+1]
  return i+1

def quick_sort_high(arr, low, high):
  if len(arr)==1:
    return arr
  if low<high:
    pivot=partition_high(arr, low, high)
    quick_sort_high(arr, low, pivot-1)
    quick_sort_high(arr, pivot+1, high)

arr=[3,12,1,4,59,26,7,53,4]
print(arr)
quick_sort_high(arr,0,len(arr)-1)
print(arr)
print()
````

# Heap Sort

### Description
* Work by dividing its input into a sorted and an unsorted region, and it iteratively shrinks the unsorted region by extracting the largest element from it and inserting it into the sorted region.


### Type
* Algorithm Strategy : 
* Algorithm Category : Selection
* Comparative Sort
* In-place (but for below examples are not-in-place, cuz I am lazy)
* Not-stable
* Efficient

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nlogn)|O(nlogn)|O(nlogn)|O(1)|

<br/>

````python
# Heap Sort
import heapq
def heap_sort(arr):
  # create a heap of the data
  h=arr.copy()
  heapq.heapify(h)
  # extract one by one from the max heap
  for i in range(len(h)):
    arr[i]=heapq.heappop(h)

heap_sort_list=[4,1,3,2,6,9,10,5,8,7]
print(heap_sort_list)
heap_sort(heap_sort_list)
print(heap_sort_list)
print()

"""# for using the data structure wrote before
def heap_sort_min(arr):
  length=len(arr)
  print(length)
  # create a min heap of the data
  h=MinHeap(length)
  h.initialize(arr)
  # extract one by one from the min heap
  for i in range(length):
    arr[i]=h.remove()"""

"""# for using the data structure wrote before
def heap_sort_max(arr):
  length=len(arr)
  print(length)
  # create a max heap of the data
  h=MaxHeap(length)
  h.initialize(arr)
  # extract one by one from the max heap
  for i in range(length-1,-1,-1):
    arr[i]=h.remove()"""
````

# Bucket Sort

### Description
* Work by distributing the elements of an array into a number of buckets.

### Type
* Algorithm Strategy : 
* Algorithm Category : Distribution
* Non-comparative Sort
* Not-in-place
* Stable (for both uniform keys and integer keys)

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n+k)|O(n+k)|O(n^2)|O(n)|
where k = number of buckets

<br/>

````python
# Bucket Sort (Integer)
def bucket_sort_integer(arr):
  # bucket array initialization
  bucket=[0]*(max(arr)+1)
  
  # storing the data by bucket
  for i in arr:
    bucket[i]=bucket[i]+1

  index=0
  for i in range(len(bucket)):
    if bucket[i]!=0:
      for j in range(bucket[i]):
        arr[index]=i
        index+=1

arr=[89,34,23,78,67,66,29,79,55,78,88,23,23]
print(arr)
bucket_sort_integer(arr)
print(arr)
````

# Counting Sort
### Description
* Work by varying bucket sort
* Use for small positive integers
* Variance of bucket sort

### Type
* Algorithm Strategy : 
* Algorithm Category : Distribution
* Non-comparative Sort
* Not-in-place
* Stable

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(n+k)|O((n+k)|O((n+k)|O((n+k)|

where $k$ is the range of the non-negative key values

<br/>

````python
# Counting Sort
def counting_sort(arr):
  number_of_count=max(arr)+1
  length=len(arr)
  output=[0]*length

  # count array initialization
  count=[0]*(number_of_count)

  # storing the count of each element 
  for i in range(0, length):
    count[arr[i]]+=1

  # storing the cumulative count
  for i in range(1, number_of_count):
    count[i]+=count[i-1]

  # place the elements in output array after finding the index of each element of original array in count array
  i=length-1
  while i>=0:
    output[count[arr[i]]-1]=arr[i]
    count[arr[i]]-=1
    i-=1

  for i in range(0, length):
    arr[i]=output[i]

arr=[3,5,1,6,7,8,3,22,3,0]
print(arr)
counting_sort(arr)
print(arr)
````

# Radix Sort

### Description
* Work by distributing elements into buckets according to their radix.

### Type
* Algorithm Strategy : 
* Algorithm Category : Distribution
* Non-comparative Sort
* Not-in-place
* Stable (for both LSD and MSD)

#### LSD Radix Sort
* Starting from the rightmost (last) digit, sort the numbers based on that digit

#### MSD Radix Sort
* Starting from the leftmost (first) digit, sort the numbers based on that digit

<br/>

|Best Time Complexity|Average Time Complexity|Worst Time Complexity|Worst Space Complexity|
|:---:|:---:|:---:|:---:|
|O(nk)|O(nk)|O(nk)|O(n+k)|
where k is length of value

<br/>

````python
def find_digit(arr):
  # use largest value to find number of digit
  digit=1
  count=max(arr)
  while count>9:
    count/=10
    digit+=1
  return digit

# Radix Sort (LSD method (Least significant digit))
def radix_sort_lsd(arr):
  cur=0
  length=len(arr)
  bucket=[[0]*length for i in range(10)] # bucket with the length same as data length, for each index in 0-9
  order=[0]*10 # store the number of data stored stored in the bucket
  digit=find_digit(arr)

  # use LSD(Least significant digit) method
  n=1
  while digit:
    for i in range(length): # arrange the bucket based on the radix
      lsd=(arr[i]//n)%10
      bucket[lsd][order[lsd]]=arr[i]
      order[lsd]+=1
    for i in range(10): # get back data from the bucket
      if order[i]!=0:
        for j in range(order[i]):
          arr[cur]=bucket[i][j]
          bucket[i][j]=0
          cur+=1
      # reset order
      order[i]=0
    n*=10
    digit-=1
    cur=0

arr=[89,34,23,78,67,100,66,29,79,55,78,1666,88,92,96,96,23]
# arr=[89,34,23,78,1784,466]
print(arr)
radix_sort_lsd(arr)
print(arr)
````
